```python
from sklearn.preprocessing import StandardScaler
# Scikit-learn, également connu sous le nom de sklearn, est une bibliothèque Python open-source pour l'apprentissage automatique (machine learning). Elle fournit des outils pour la classification, la régression, le clustering, la sélection de modèles, la préparation des données et l'évaluation des modèles.
# Scikit-learn est construit sur NumPy, SciPy et matplotlib, ce qui en fait une bibliothèque très performante pour l'apprentissage automatique en Python. Elle est conçue pour être simple et efficace, tout en offrant une grande flexibilité pour les utilisateurs.
# Scikit-learn est largement utilisé dans les domaines de la science des données, de l'analyse statistique, de la bioinformatique, de l'ingénierie et de la recherche. Elle offre une grande variété d'algorithmes d'apprentissage automatique, tels que les arbres de décision, les SVM, les réseaux de neurones, les méthodes ensemblistes, etc.

scaler = StandardScaler()
scaler.fit(iris)
iris_cr=scaler.transform(iris)
print(iris_cr)

# Ce code effectue une mise à l'échelle (scaling) des données de `iris` en utilisant la classe `StandardScaler` de la bibliothèque Scikit-learn (sklearn).
# La mise à l'échelle des données est une étape importante dans le prétraitement des données pour l'apprentissage automatique (machine learning), car elle permet de normaliser les données et de les rendre comparables entre elles. La classe `StandardScaler` est l'un des outils les plus couramment utilisés pour la mise à l'échelle des données, car elle transforme les données en les centrant autour de zéro et en les réduisant à une variance unitaire.
# Le code commence par importer la classe `StandardScaler` à partir du module `preprocessing` de la bibliothèque Scikit-learn. Ensuite, il crée une instance de la classe `StandardScaler` appelée `scaler`.
# La méthode `fit()` de `scaler` est ensuite appelée pour calculer les moyennes et les écarts-types des caractéristiques de `iris`. Ces statistiques sont utilisées pour normaliser les données dans la méthode `transform()`.
# La méthode `transform()` de `scaler` est ensuite utilisée pour appliquer la transformation de mise à l'échelle aux données de `iris`. Le nouveau dataframe normalisé est stocké dans la variable `iris_cr`. Enfin, le code imprime le nouveau dataframe normalisé en utilisant la fonction `print()`.
```
